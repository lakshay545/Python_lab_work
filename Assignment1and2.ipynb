 Q1. The prices of five bikes in a shop are: 
₹12,000, ₹15,000, ₹18,000, ₹20,000, ₹50,000 
(a) What is the mean price of a bike? (2 Marks) 
(b) What is the median price of a bike? (2 Marks) 
(c) If the shop owner wants to make the 

```python
prices = [12000, 15000, 18000, 20000, 50000]

total = 0
for p in prices:
    total = total + p
mean = total / len(prices)

prices.sort()
n = len(prices)
middle = n // 2
median = prices[middle]

print("Mean price =", mean)
print("Median price =", median)

if mean > median:
    print("Shopkeeper should use Median because it looks cheaper.")
else:
    print("Shopkeeper should use Mean.")
```

Q2.	 Describe the relationship between a histogram's  shape and the values of its mean, median, and 
mode in a left-skewed distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
import statistics

rng = np.random.default_rng(seed=42)
sample_size = 1000
right_skew = rng.exponential(scale=1.0, size=sample_size)
left_skewed = 10 - right_skew  

mean_val = np.mean(left_skewed)
median_val = np.median(left_skewed)

counts, bin_edges = np.histogram(left_skewed, bins=30)
max_bin_index = np.argmax(counts)
mode_estimate = (bin_edges[max_bin_index] + bin_edges[max_bin_index + 1]) / 2

print(f"Mean ≈ {mean_val:.4f}")
print(f"Median ≈ {median_val:.4f}")
print(f"Mode (histogram estimate) ≈ {mode_estimate:.4f}")

plt.figure(figsize=(10,5))
plt.hist(left_skewed, bins=30, edgecolor='black')
plt.axvline(mean_val, linestyle='--', linewidth=2, label=f"Mean = {mean_val:.2f}")
plt.axvline(median_val, linestyle='-.', linewidth=2, label=f"Median = {median_val:.2f}")
plt.axvline(mode_estimate, linestyle=':', linewidth=2, label=f"Mode ≈ {mode_estimate:.2f}")
plt.title("Histogram of a Left-Skewed Sample (tail to the LEFT)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

Q3.	 The following data represents the exam scores  (out of 100) for 10 students: 
85, 92, 78, 90, 85, 67, 88, 95, 85, 72 
(a) Calculate the mean, median, and mode for this data. 
(b) Based on your values from part (a), describe the shape of the distribution 
(symmetrical, skewed left, or skewed right). Justify your answer. 
(c) Find the Five-Number Summary for this data.

```python
import numpy as np
import statistics as stats
import matplotlib.pyplot as plt

data = [85, 92, 78, 90, 85, 67, 88, 95, 85, 72]

mean = np.mean(data)
median = np.median(data)
mode = stats.mode(data)

print("(a) Mean:", mean)
print("(a) Median:", median)
print("(a) Mode:", mode)

if mean < median and median == mode:
    shape = "Slightly Left-Skewed (Negatively skewed)"
elif mean > median and median == mode:
    shape = "Slightly Right-Skewed (Positively skewed)"
else:
    shape = "Approximately Symmetrical"
print("(b) Shape of distribution:", shape)

minimum = np.min(data)
q1 = np.percentile(data, 25)
q2 = np.median(data)  
q3 = np.percentile(data, 75)
maximum = np.max(data)

five_num_summary = (minimum, q1, q2, q3, maximum)
print("(c) Five-Number Summary (Min, Q1, Median, Q3, Max):")
print(five_num_summary)

plt.hist(data, bins=6, color="lightblue", edgecolor="black")
plt.axvline(mean, color="red", linestyle="--", label=f"Mean = {mean:.1f}")
plt.axvline(median, color="green", linestyle="-.", label=f"Median = {median}")
plt.axvline(mode, color="purple", linestyle=":", label=f"Mode = {mode}")
plt.title("Histogram of Exam Scores")
plt.xlabel("Scores")
plt.ylabel("Frequency")
plt.legend()
plt.show()

plt.boxplot(data, vert=False, patch_artist=True, boxprops=dict(facecolor="lightyellow"))
plt.title("Boxplot of Exam Scores (Five-Number Summary)")
plt.xlabel("Scores")
plt.show()
```

Q4	 . Differentiate between a population parameter and  a sample statistic with an example of each.

```python
import numpy as np
import matplotlib.pyplot as plt

population = np.random.normal(loc=170, scale=10, size=10000) 
population_mean = np.mean(population)  
print("Population mean (parameter):", population_mean)

sample = np.random.choice(population, size=100, replace=False)
sample_mean = np.mean(sample)  
print("Sample mean (statistic):", sample_mean)

plt.figure(figsize=(10,6))
plt.hist(population, bins=30, alpha=0.6, label='Population', color='skyblue')

plt.axvline(population_mean, color='blue', linestyle='dashed', linewidth=2, label=f'Population Mean = {population_mean:.2f}')

plt.axvline(sample_mean, color='red', linestyle='solid', linewidth=2, label=f'Sample Mean = {sample_mean:.2f}')

plt.title("Population vs Sample Mean")
plt.xlabel("Height (cm)")
plt.ylabel("Frequency")
plt.legend()
plt.show()

```

Q.6	 The following data represents the exam scores  (out of 100) for 10 students: 
85, 92, 78, 90, 85, 67, 88, 95, 85, 72 
(a) Calculate the mean, median, and mode for this data. 
(b) Based on your values from part (a), describe the shape of the distribution 
(symmetrical, skewed left, or skewed right). Justify your answer. 
(c) Find the Five-Number Summary for this data.

```python
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

scores = [85, 92, 78, 90, 85, 67, 88, 95, 85, 72]

mean = np.mean(scores)

median = np.median(scores)

mode_data = Counter(scores)
mode = mode_data.most_common(1)[0][0]

minimum = np.min(scores)
q1 = np.percentile(scores, 25)
q3 = np.percentile(scores, 75)
maximum = np.max(scores)

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
print("Five-number summary:", (minimum, q1, median, q3, maximum))

if mean < median and median == mode:
    shape = "Slightly Left-Skewed"
elif mean > median and median == mode:
    shape = "Slightly Right-Skewed"
else:
    shape = "Approximately Symmetric"
print("Shape of Distribution:", shape)

plt.figure(figsize=(7,4))
plt.boxplot(scores, vert=False, patch_artist=True,
            boxprops=dict(facecolor="lightblue", color="blue"),
            medianprops=dict(color="red", linewidth=2),
            whiskerprops=dict(color="blue"),
            capprops=dict(color="blue"))
plt.title("Boxplot of Exam Scores")
plt.xlabel("Scores")
plt.show()
```

Q7. Graphical Representation 
A quick survey of 20 students about their preferred mode of learning yielded the following data: 
(H: Hybrid, O: Online, F: In-Person/Face-to-Face) 
Raw Data: F, H, O, F, F, O, H, F, O, F, H, F, F, F, O, H, O, F, H, F 
(a) Organize this raw categorical data into a Frequency Table. 
(b) Based on your table, create a Bar Chart to represent this data. Remember to label your axes and 
give the chart a title. 
(c) Based on the chart, what is the most popular learning mode? What percentage of students prefer 
it? 
(d) Why is a Histogram not a suitable graph for this type of data?

```python
import matplotlib.pyplot as plt

data = ["F", "H", "O", "F", "F", "O", "H", "F", "O", "F", 
        "H", "F", "F", "F", "O", "H", "O", "F", "H", "F"]

freq_F = data.count("F")
freq_H = data.count("H")
freq_O = data.count("O")

frequency_table = {
    "Face-to-Face (F)": freq_F,
    "Hybrid (H)": freq_H,
    "Online (O)": freq_O
}

print("Frequency Table:")
for mode, freq in frequency_table.items():
    print(f"{mode}: {freq}")
print("Total:", len(data))

modes = list(frequency_table.keys())
frequency = list(frequency_table.values())

plt.bar(modes, frequency, color=["skyblue", "orange", "green"])
plt.xlabel("Learning Mode")
plt.ylabel("Number of Students")
plt.title("Preferred Mode of Learning among Students")
plt.show()

most_popular_mode = max(frequency_table, key=frequency_table.get)
most_popular_count = frequency_table[most_popular_mode]
percentage = (most_popular_count / len(data)) * 100

print("\nMost popular learning mode:", most_popular_mode)
print("Percentage of students preferring it:", percentage, "%")

print("\nNote: A Histogram is not suitable here because it is used for continuous numerical data,")
print("but our data (F, H, O) is categorical, so a Bar Chart is appropriate.")
```

was blue?

Q2	 . Suppose a laboratory test on a blood sample yields  two possible results, positive or negative. 
According to the industry reports, the blood sample 95% of the people with a particular disease 
yield positive results. But 2% of the people without the disease also give positive results (false 
positive). 1% of the total population is infected by the disease. Determine the probability that a 
person chosen randomly from the population will have disease given that the blood sample of the 
person tests positive?

```python
P_D = 0.01      
P_notD = 1 - P_D  

P_T_given_D = 0.95  
P_T_given_notD = 0.02  

P_D_given_T = (P_T_given_D * P_D) / ((P_T_given_D * P_D) + (P_T_given_notD * P_notD))

print("Probability that a person has the disease given a positive test:", P_D_given_T)
```

Q4.	 In a group of 50 students, 30 study Physics, 25  study Chemistry, and 15 study both. Draw a Venn 
diagram to represent this and find the probability that a randomly selected student studies only 
Physics. 

```python

import matplotlib.pyplot as plt
from matplotlib.patches import Circle

# Given data
total = 50
physics = 30
chemistry = 25
both = 15

# Calculations
only_physics = physics - both
only_chemistry = chemistry - both
neither = total - (only_physics + only_chemistry + both)
prob_only_physics = only_physics / total

print("Only Physics =", only_physics, "students")
print("Only Chemistry =", only_chemistry, "students")
print("Both =", both, "students")
print("Neither =", neither, "students")
print(f"Probability (only Physics) = {only_physics}/{total} = {prob_only_physics:.2f} = {prob_only_physics*100:.0f}%")

# Draw a simple Venn-like diagram
fig, ax = plt.subplots(figsize=(6,5))
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)
ax.set_aspect('equal')
ax.axis('off')

# Two overlapping circles
c1 = Circle((4,5.5), 3, alpha=0.3, edgecolor='black', linewidth=1.2)
c2 = Circle((6,5.5), 3, alpha=0.3, edgecolor='black', linewidth=1.2)
ax.add_patch(c1)
ax.add_patch(c2)

# Text for regions
ax.text(3.2, 5.5, f"Only P\n{only_physics}", ha='center', va='center', fontsize=12)
ax.text(6.8, 5.5, f"Only C\n{only_chemistry}", ha='center', va='center', fontsize=12)
ax.text(5, 5.5, f"Both\n{both}", ha='center', va='center', fontsize=12)
ax.text(5, 2, f"Neither\n{neither}", ha='center', va='center', fontsize=12)

# Titles
plt.title("Venn Diagram (Physics & Chemistry)\nCounts in each region", fontsize=14)
plt.show()
```
